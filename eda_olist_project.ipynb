```python
import pandas as pd
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
sellers = pd.read_csv("olist_sellers_dataset.csv")
category_translation = pd.read_csv("product_category_name_translation.csv")
```

```python
# Data Cleaning & Preparation
#-----------------------------------------------------------------------------------------
```

```python
# Merge orders + customers
df = pd.merge(orders, customers, on='customer_id', how='left')

# Merge order_items
df = pd.merge(df, order_items, on='order_id', how='left')

# Merge payments
df = pd.merge(df, payments, on='order_id', how='left')

# Merge reviews
df = pd.merge(df, reviews[['order_id', 'review_score']], on='order_id', how='left')

# Merge products
df = pd.merge(df, products[['product_id', 'product_category_name']], on='product_id', how='left')

# Merge category translation
df = pd.merge(df, category_translation, on='product_category_name', how='left')

# Rename English category name column
df.rename(columns={'product_category_name_english': 'category_english'}, inplace=True)
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#Handle Missing Values
# Check nulls
df.isnull().sum()


# Fill missing category names
df['category_english'] = df['category_english'].fillna("Unknown")

# Drop rows with critical nulls
df.dropna(subset=['order_purchase_timestamp', 'order_delivered_customer_date'], inplace=True)
```

```python
#Convert Date Columns to datetime Format
date_cols = ['order_purchase_timestamp', 'order_approved_at', 
             'order_delivered_carrier_date', 'order_delivered_customer_date', 
             'order_estimated_delivery_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

```python
#Create New Columns
# Order Month
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Delivery Days
df['delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# Total Price
df['total_price'] = df['price'] + df['freight_value']

df[['order_id', 'order_month', 'delivery_days', 'total_price']].head()

```

```python
# Define churned customers (inactive > X days)
#We'll define churned users as those who haven‚Äôt placed any order in the last 90 days of the dataset.
# Latest order date
max_date = df['order_purchase_timestamp'].max()

# Last order per customer
last_purchase = df.groupby('customer_unique_id')['order_purchase_timestamp'].max()

# Days since last order
days_inactive = (max_date - last_purchase).dt.days

# Define churned (inactive > 90 days)
churned_customers = days_inactive[days_inactive > 90]
active_customers = days_inactive[days_inactive <= 90]

print(f"Total Churned Customers: {len(churned_customers)}")
print(f"Total Active Customers: {len(active_customers)}")
```

```python
# Retention rate month-on-month
# Get first order month of each customer
first_order = df.groupby('customer_unique_id')['order_purchase_timestamp'].min()
first_order_month = first_order.dt.to_period('M')

# Add to main df
df['first_order_month'] = df['customer_unique_id'].map(first_order_month)

# Create user activity table
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
cohort_table = df.groupby(['first_order_month', 'order_month'])['customer_unique_id'].nunique().unstack(fill_value=0)

# Retention = later month / first month
retention_rate = cohort_table.divide(cohort_table.iloc[:,0], axis=0).round(2)

# Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(retention_rate, annot=True, fmt='.0%', cmap='YlGnBu')
plt.title('Customer Retention Rate (%) by Cohort Month')
plt.xlabel('Order Month')
plt.ylabel('First Order Month')
plt.tight_layout()
plt.show()

```

```python
#RFM Segmentation
```

```python
#Goal:Segment customers into groups like Loyal, At Risk, Champions, etc. based on:
#Recency = How recently they purchased
#Frequency = How often they purchased
#Monetary = How much they spent
```

```python
#Step 1: RFM Base Table Creation
```

```python
# Reference date (latest date in dataset)
today = df['order_purchase_timestamp'].max()

# Group by customer
rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (today - x.max()).days,  # Recency
    'order_id': 'nunique',                                          # Frequency
    'total_price': 'sum'                                            # Monetary
}).reset_index()

# Rename columns
rfm.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
#Step 2: Assign R, F, M Scores (1 to 5)
#We‚Äôll use quintiles to assign scores.
# R Score: lower is better (recent)
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1]).astype(int)
# F Score: higher is better
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
# M Score: higher is better
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5]).astype(int)
# Combine RFM Score
rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm.head()
```

```python
#Step 3: Define Customer Segments
def segment_customer(row):
    if row['R_score'] >= 4 and row['F_score'] >= 4:
        return 'Champions'
    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal'
    elif row['R_score'] >= 4 and row['F_score'] <= 2:
        return 'New Customers'
    elif row['R_score'] <= 2 and row['F_score'] >= 4:
        return 'At Risk'
    elif row['R_score'] == 1 and row['F_score'] == 1:
        return 'Lost'
    else:
        return 'Need Attention'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
rfm.head()
```

```python
#Segment Count Visualization
# Count of each segment
segment_counts = rfm['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Count']

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=segment_counts, x='Count', y='Segment', hue='Segment', palette='Set2', legend=False)
plt.title('Customer Segmentation using RFM')
plt.xlabel('Number of Customers')
plt.ylabel('Segment')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(segment_counts['Count'], labels=segment_counts['Segment'], autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('RFM Segment Distribution')
plt.tight_layout()
plt.show()
```
